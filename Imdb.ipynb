```python
sns.heatmap(data.isnull(),cmap='seismic')
plt.show()
```

Drop all Rows with the Missing Values

```python
data = data.dropna(axis=0)
```

Heatmap after dropping missing values

```python
sns.heatmap(data.isnull(),cmap = 'cividis')
plt.show()
```

Check For Duplicate Data

```python
duplicate_data = data.duplicated().any()
print('Are there any Duplicated values in the data?','Yes' if duplicate_data else 'No')
```

Get Overall Statistics About The DataFrame

```python
data.describe()
```

Display Title of The Movie Having Runtime >= 150 Minutes

```python
Result = data[data['Runtime (Minutes)']>=150]['Title']
print(Result)
print('Movies with Runtime >= 150 are',Result.count(),'in total.')
```

Year There with the Highest Voting

```python
sns.barplot(x='Year',y='Votes',data=data, palette='crest', hue='Year', legend=False)
plt.title('Votes By Year')
plt.show()

max_votes_year = data.loc[data['Votes'].idxmax(), 'Year']
max_votes = data['Votes'].max()

print(f"\nThe year with the highest votes is {max_votes_year} with {max_votes} votes.")
```

Year with The Highest Revenue

```python
sns.barplot(x='Year', y='Revenue (Millions)', data=data, palette='muted', hue='Year', legend=False)
plt.title("Revenue By Year")
plt.show()

max_revenue_year = data.loc[data['Revenue (Millions)'].idxmax(), 'Year']
max_revenue = data['Revenue (Millions)'].max()

print(f"\nThe year with the highest revenue is {max_revenue_year} with {max_revenue} million.")
```

Average Rating For Each Director

```python
data.groupby('Director')['Rating'].mean().sort_values(ascending=False)
```

Top 10 Lengthy Movies Title

```python
le =data.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']].set_index('Title')
```

```python
sns.barplot(x='Runtime (Minutes)', y=le.index, data=le, palette='deep',hue=le.index, legend=False)
plt.title('Top 5 Lengtly Movies')
plt.show()
```

Number of Movies Per Year

```python
sns.countplot(x='Year',data=data, hue='Year',legend=False, palette='viridis')
plt.title("Number of Movies Per Year")
```
