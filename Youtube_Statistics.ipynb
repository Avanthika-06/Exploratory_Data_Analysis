 calculatation of the mean number of subscribers for each category in the dataset

```python
categories_subscribers_mean = df.groupby("category")["subscribers"].mean()
categories_subscribers_mean
```

Sorting the values

```python
categories_subscribers_mean.sort_values(ascending = False)
```

Removing all the null values

```python
df['country'] = df['country'].fillna('Unknown')
df['category'] = df['category'].fillna('Unknown')
df['abbreviation'] = df['abbreviation'].fillna('Unknown')
df['channel_type'] = df['channel_type'].fillna('Unknown')
df['country_rank'] = df['country_rank'].fillna(df['country_rank'].mean())
df['channel_type_rank'] = df['channel_type_rank'].fillna(df['channel_type_rank'].mean())
df['video_views_for_the_last_30_days'] = df['video_views_for_the_last_30_days'].fillna(df['video_views_for_the_last_30_days'].mean())
df['subscribers_for_last_30_days'] = df['subscribers_for_last_30_days'].fillna(df['subscribers_for_last_30_days'].mean())
df['video_views_rank'] = df['video_views_rank'].fillna('0')
df['population'] = df['population'].fillna(df['population'].mean())
df['unemployment_rate'] = df['unemployment_rate'].fillna(df['unemployment_rate'].mean())
df['urban_population'] = df['urban_population'].fillna(df['urban_population'].mean())
df['latitude'] = df['latitude'].fillna(df['latitude'].mean())
df['longitude'] = df['longitude'].fillna(df['longitude'].mean())
```

```python
df.isnull().sum()
```

```python
df['created_year'] = df['created_year'].fillna('Unknown')
df['created_month'] = df['created_month'].fillna('Unknown')
df['created_date'] = df['created_date'].fillna('Unknown')
```

```python
df.isnull().sum().sum()
```

Mean Subscribers by Category

```python
categories_subscribers_mean.plot(kind = "bar")
plt.title('Mean Subscribers by Category')
plt.xlabel('Category')
plt.ylabel('Mean Subscribers')
plt.show()
```

Channel Type Distribution

```python
channel_type_counts = df['channel_type'].value_counts()
channel_type_counts
channel_type_counts.head().plot(kind="pie", autopct="%1.1f%%")
plt.title("Channel Type Distribution")
plt.show()
```

Subscribers vs Video views

```python
plt.scatter(df['subscribers'],df['video_views'])
plt.title('Subscribers vs Video views')
plt.xlabel('Subscribers')
plt.ylabel('Video views')
plt.show()
```

```python
country_subscribers_total = df.groupby('country')["subscribers"].sum().sort_values(ascending = False)
country_subscribers_total
```

Total Subscribers by Country

```python
country_subscribers_total.head(10).plot(kind = "bar", color=sns.color_palette('muted', n_colors=10))
plt.title("Total Subscribers by Country")
plt.xlabel("Country")
plt.ylabel("Total Subscribers")
plt.show()
```

Top categories by views or subscribers

```python
top_categories = df.groupby('category')['video_views'].sum().sort_values(ascending=False).head(10)
top_categories.plot(kind='bar', color='orange')
plt.title('Top 10 Categories by Views')
plt.ylabel('Total Views')
plt.show()
```

```python
invalid_dates = df[df['created_date'] == 'Unknown']
df['created_date'] = pd.to_datetime(df['created_date'], errors='coerce')
df['created_date'] = pd.to_datetime(df['created_date'], errors='coerce').fillna(pd.Timestamp('2020-01-01'))
df = df[pd.to_datetime(df['created_date'], errors='coerce').notna()]
df.groupby(df['created_date'].dt.year)['video_views'].sum().plot(kind='line')
plt.title('Video Views Over Time')
plt.xlabel('Year')
plt.ylabel('Total Views')
plt.show()
```
